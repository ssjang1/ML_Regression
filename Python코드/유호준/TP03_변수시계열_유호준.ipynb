```python
import numpy as np
from sklearn.decomposition import PCA

# 데이터 준비


# PCA 모델 생성 및 학습
pca = PCA(n_components=2)
pca.fit(ttt)

# 주성분 변환
X_transformed = pca.transform(ttt)

# 주성분 확인
components = pca.components_

# 변환된 데이터 출력
print("Original Data:\n", X)
print("Transformed Data:\n", X_transformed)
print("Components:\n", components)
```

```python
주어진 데이터에 대한 PCA 결과를 해석해보겠습니다.

Original Data: 주어진 데이터는 2개의 열로 구성되어 있습니다. 첫 번째 열은 상수 항으로 모두 1로 구성되어 있고, 두 번째 열은 '수관폭평균'을 나타냅니다. 총 2207개의 데이터 포인트가 있습니다.

Transformed Data: PCA를 적용한 결과, 주어진 데이터가 2차원으로 변환되었습니다. 두 번째 열은 모두 0으로 표시되고, 첫 번째 열은 변환된 주성분 값을 나타냅니다. 변환된 데이터는 2차원 공간상의 좌표로 표현됩니다. 예시에서는 모든 데이터의 두 번째 열이 0이므로 변환된 데이터는 첫 번째 열 값에 의존합니다.

Components: PCA를 통해 추출된 주성분은 2차원 공간에서의 벡터로 나타내어집니다.
주성분은 변환된 데이터의 축을 나타내며, 이 예시에서는 첫 번째 열이 0, 두 번째 열이 1인 벡터와 첫 번째 열이 1, 두 번째 열이 0인 벡터가 주성분으로 추출되었습니다. 이는 원본 데이터의 주요 변동성을 나타내는 방향을 의미합니다.

위의 결과를 종합하여 해석하면, 주어진 데이터에서 변환된 데이터는 2차원 평면상에서 하나의 축에 대해 변동성을 가지지 않고 다른 축에 대해서만 변동성을 가지고 있음을 알 수 있습니다. 주성분 벡터로 나타내면 변환된 데이터는 모두 첫 번째 열 값에 의존하며, 두 번째 열은 변화하지 않습니다. 따라서 두 번째 열인 '수관폭평균'은 변환된 데이터에 영향을 미치지 않는다고 해석할 수 있습니다.






```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# y 데이터프레임에서 착과량 정보를 가져옴
y_values = y.values[0]

# y 데이터프레임에 해당하는 샘플만 선택
ssnd_selected = ssnd.iloc[:, :len(y_values)]

# y_train의 샘플 수를 X_train과 동일하게 맞추기 위해 샘플링
X_train, X_temp, y_train, y_temp = train_test_split(ssnd_selected, y_values[:len(ssnd_selected)], train_size=0.7, random_state=3)

# 데이터셋을 훈련 세트와 테스트 세트로 나눔
X_test, X_val, y_test, y_val = train_test_split(X_temp, y_temp, test_size=0.3, random_state=3)

# 결과 확인
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_val shape:", X_val.shape)
print("y_val shape:", y_val.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# 랜덤 포레스트 모델 생성
rf_model = RandomForestRegressor(random_state=3)

# 모델 훈련
rf_model.fit(X_train, y_train)

# 검증 세트로 예측
val_predictions = rf_model.predict(X_val)

# 검증 세트의 평균 제곱 오차 계산
val_mse = mean_squared_error(y_val, val_predictions)
val_rmse = val_mse ** 0.5

# 결과 출력
print("Validation RMSE:", val_rmse)
```

```python
#Validation RMSE (Root Mean Squared Error)는 모델의 예측값과 실제값 간의 차이를 나타내는 지표입니다. RMSE 값이 작을수록 모델의 예측이 실제값과 잘 일치한다는 것을 의미합니다.

#해당 RMSE 값이 185.8957281842239인 경우, 예측값과 실제값 사이의 평균적인 차이가 약 185.8957281842239 정도라는 것을 의미합니다.
#이 값이 크다면 모델의 예측이 실제값과 상당히 다르다는 것을 나타내며, 작을수록 예측의 정확도가 높다고 해석할 수 있습니다.

#RMSE는 주로 회귀 모델의 성능을 평가하는 지표로 사용되며, 낮은 RMSE 값을 가지는 모델이 일반적으로 성능이 좋다고 판단됩니다.
#따라서, 해당 모델의 경우 185.8957281842239의 RMSE 값을 가지므로 예측 성능이 상대적으로 좋지 않다고 판단할 수 있습니다.
```

```python
import xgboost as xgb
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# XGBoost 모델 초기화
model = xgb.XGBRegressor()

# 모델 훈련
model.fit(X_train, y_train)

# 검증 세트로 예측
y_pred = model.predict(X_val)

# 실제 값과 예측 값의 비교를 위해 데이터프레임으로 변환
df_pred = pd.DataFrame({'Actual': y_val, 'Predicted': y_pred})

# 실제 값과 예측 값의 시각화
plt.figure(figsize=(12, 6))
df_pred.plot(kind='line')
plt.title('Actual vs Predicted')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend(loc='upper right')
plt.show()

# RMSE 계산
rmse = mean_squared_error(y_val, y_pred, squared=False)
print("Validation RMSE:", rmse)


```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd

# 데이터 준비 (Pandas DataFrame)


# 데이터프레임을 NumPy 배열로 변환
X = scaled_df.values

# K-means 클러스터링 모델 생성 및 학습
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)

# 클러스터링 결과 확인
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# 클러스터링 결과 출력 및 시각화
for i in range(len(X)):
    print("Data:", X[i, 0], "Label:", labels[i])

print("Centroids:", centroids)

# 데이터 포인트 시각화
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', color='red', s=200)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("K-means Clustering")
plt.show()
```
