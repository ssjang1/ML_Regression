```python
import pandas as pd
import numpy as np
```

```python
path = '/content/drive/MyDrive/ P - value/proj_3/데이터 파일/전처리데이터 (전후)/'
date_df = pd.read_csv(path + 'X_scaled_date.csv')
make_csv_name = 'X_scaled_Trend.csv'
```

```python
target_columns = date_df.filter(regex="새순").columns
sprout_date_df = date_df[target_columns].T
sprout_date_df.shape
```

```python
target_columns = date_df.filter(regex="엽록소").columns
chlorophyll_date_df = date_df[target_columns].T
chlorophyll_date_df.shape
```

```python
sprout_slope = []
sprout_intercept = []

for item, trend in sprout_date_df.items():
  
    x = np.arange(len(trend)) 
    y = np.array(trend) 
    coefficients = np.polyfit(x, y, deg=1)  # 1차 다항식으로 추세선 근사화
    slope, intercept = coefficients  # 추세선의 기울기는 1차 다항식의 계수
    sprout_slope.append(slope)
    sprout_intercept.append(intercept)
```

```python
chlorophyll_slope = []
chlorophyll_intercept = []

for item, trend in chlorophyll_date_df.items():
  
    x = np.arange(len(trend)) 
    y = np.array(trend) 
    coefficients = np.polyfit(x, y, deg=1)  # 1차 다항식으로 추세선 근사화
    slope, intercept = coefficients  # 추세선의 기울기는 1차 다항식의 계수
    chlorophyll_slope.append(slope)
    chlorophyll_intercept.append(intercept)
```

```python
result_df = pd.DataFrame({'sprout_slope' : sprout_slope,
                         'sprout_intercept' : sprout_intercept,
                         'chlorophyll_slope' : chlorophyll_slope,
                         'chlorophyll_intercept' : chlorophyll_intercept})
```

```python
result_df.shape
```

```python
result_df.to_csv(path + make_csv_name ,index=False)
```
