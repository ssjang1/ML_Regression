# 이상치 처리

- IQR을 이용한 방법

```python
import numpy as np
Q1 = X.apply(lambda x: x.quantile(0.25))
Q3 = X.apply(lambda x: x.quantile(0.75))
IQR=Q3-Q1
outliers_IQR = X.apply(lambda x: (x < Q1[x.name] - IQR[x.name] * 1.5) | (x > Q3[x.name] + IQR[x.name] * 1.5))
```

```python
# 열별 이상치 확인
outliers_IQR.sum()
```

```python
# 각 행별로 True의 개수 계산
outliers_IQR_count = outliers_IQR.sum(axis=1)

# 정렬하여 데이터프레임 생성
sorted_IQR = pd.DataFrame({'outliers_IQR_count': outliers_IQR_count}).sort_values(by='outliers_IQR_count', ascending=False)

# 결과 출력
print(sorted_IQR.head(50))
```

- ESD를 이용한 방법(평균,표준편차의 3배수)

* 이 방법 사용했음




```python
outliers_ESD=(X<X.mean()-3*X.std()) | (X>X.mean()+3*X.std())
outliers_ESD.sum()
```

```python
# 각 행별로 True의 개수 계산
outliers_ESD_count = outliers_ESD.sum(axis=1)

# 정렬하여 데이터프레임 생성
#sorted_ESD = pd.DataFrame({'outliers_ESD_count': outliers_ESD_count}).sort_values(by='outliers_ESD_count', ascending=False)

# 결과 출력
#print(sorted_ESD.head(50))
# 7개의 값이 이상치 2063,420,1227,2130,629,1696,480

out_index = outliers_ESD_count.index[outliers_ESD_count >= 10]
out_index
```
