# PCA분석

```python
import numpy as np
from sklearn.decomposition import PCA

# 주성분 분석 객체 생성
pca = PCA()

# 주성분 분석 수행
pca.fit(X_scaled_concat)

# 주성분의 설명 분산 비율 추출
explained_variance_ratio = pca.explained_variance_ratio_

# scree plot 그리기
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o')
plt.xlabel('주성분 개수')
plt.ylabel('주성분 설명 분산 비율')
plt.title('Scree Plot')
plt.show()
```

```python
# 주성분 분석 객체 생성
pca = PCA(n_components=5)  # 주성분 개수를 5로 설정

# 주성분 분석 수행
X_pca = pca.fit_transform(X_scaled_concat)

# 주성분 분석 결과 확인
explained_variance_ratio = pca.explained_variance_ratio_
print("주성분 설명 분산 비율:", explained_variance_ratio)
print('총 설명 분산 비율:', round(sum(explained_variance_ratio),2))

# 주성분 분석 결과 데이터프레임 생성
df_pca = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5'])
```

```python
mse_values=[]
r2_values=[]
row_name=['선형','다항','릿지','라쏘','엘라스틱넷','의사결정트리','서포트벡터머신']
```

# 데이터 분할

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_pca, y_scaled, test_size=0.3, random_state=3)
```

# 선형회귀

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# 모델 객체 생성
model = LinearRegression()

# 모델 피팅
model.fit(X_train, y_train)

# 예측
y_pred = model.predict(X_test)

# 회귀 모델의 정확성 분석
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error (MSE):", mse)
print("R-squared (R^2):", r2)
mse_values.append(mse)
r2_values.append(r2)
```

# k-means

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# 엘보우 커브를 사용하여 최적의 군집 개수 결정
k_values = range(2, 11)  # 시도할 군집 개수 범위
inertia_values = []  # 군집 내 제곱합(Inertia)을 저장할 리스트

for k in k_values:
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=3)
    kmeans.fit(X_pca)
    inertia_values.append(kmeans.inertia_)

# 엘보우 커브 시각화
plt.plot(k_values, inertia_values, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Curve')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
# 실루엣 계수를 사용하여 최적의 군집 개수 결정
silhouette_scores = []  # 실루엣 계수를 저장할 리스트

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=3, n_init=10)
    kmeans.fit(X_pca)
    labels = kmeans.labels_
    silhouette_avg = silhouette_score(X_pca, labels)
    silhouette_scores.append(silhouette_avg)

# 실루엣 계수 시각화
plt.plot(k_values, silhouette_scores, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Method')
plt.show()
```

```python
from sklearn.cluster import DBSCAN

# DBSCAN 모델 생성 및 학습
dbscan = DBSCAN(eps=0.5, min_samples=5)  # eps: 이웃 포인트 간의 거리 임계값, min_samples: 코어 포인트로 인정되기 위한 이웃 포인트의 최소 개수
dbscan.fit(df_pca)

# 군집 레이블 출력
labels = dbscan.labels_
print("Cluster Labels:", labels)
```

# KNN 군집분석

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# 학습 데이터와 테스트 데이터로 나누기
X_train, X_test, y_train, y_test = train_test_split(df_pca, y_scaled, test_size=0.3, random_state=3)

# KNN 모델 생성 및 학습
knn = KNeighborsClassifier(n_neighbors=3)  # 이웃의 개수를 3으로 설정
knn.fit(X_train, y_train)

# 테스트 데이터에 대한 예측
y_pred = knn.predict(X_test)

# 정확도 평가
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
