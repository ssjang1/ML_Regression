```python
phh = phh_y_scaled.loc[:, '착과량(int)'].copy()
```

```python
phh = pd.concat([phh, phh_X_scaled[['수고(m)', '수관폭평균']]], axis=1)
phh = pd.concat([phh, phh_X_scaled_date[['2022-09-01 새순', '2022-10-01 새순', '2022-09-01 엽록소', '2022-10-01 엽록소']]], axis=1)
```

```python
phh
```

```python
from sklearn.model_selection import train_test_split    
from sklearn.ensemble import RandomForestClassifier
```

```python
# X, y 나눠주기
X = phh.drop('착과량(int)', axis=1)  # 변수 열들을 X로 설정
y = phh['착과량(int)']  # 타겟 열을 y로 설정

# Train-Test 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)

# Random Forest Classifier 모델 생성 및 학습
rf = RandomForestClassifier(random_state=3)
rf.fit(X_train, y_train)

# 테스트 데이터에 대한 예측
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

# 정확도를 계산해보자
accuracy = accuracy_score(y_test, y_pred)

print("정확도 : ", accuracy)
```

```python
import numpy as np

# 예측값과 실제값의 차이 계산
errors = np.abs(y_pred - y_test)

# 오차 범위 통계량 계산
mean_error = np.mean(errors)
max_error = np.max(errors)
min_error = np.min(errors)
median_error = np.median(errors)
std_error = np.std(errors)

print("Mean Error:", mean_error)
print("Max Error:", max_error)
print("Min Error:", min_error)
print("Median Error:", median_error)
print("Standard Deviation of Error:", std_error)
```
