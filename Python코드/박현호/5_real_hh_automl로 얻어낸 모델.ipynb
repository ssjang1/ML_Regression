```python
phh = phh_y_scaled.loc[:, '착과량(int)'].copy()
```

```python
phh = pd.concat([phh, phh_X_scaled], axis=1)
phh = pd.concat([phh, phh_X_scaled_date], axis=1)
```

```python
phh
```

```python
# X, y 나눠주기
X = phh.drop('착과량(int)', axis=1)  # 변수 열들을 X로 설정
y = phh['착과량(int)']  # 타겟 열을 y로 설정
```

```python
!pip install mljar-supervised
```

```python
from supervised.automl import AutoML
from sklearn.model_selection import train_test_split

# Train-Test 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)

# AutoML 모델 생성 및 학습
automl = AutoML(mode="Compete", eval_metric='rmse', algorithms=['Xgboost', 'Extra Trees', 'LightGBM'], stack_models=True)
automl.fit(X_train, y_train)

# 테스트 데이터에 대한 예측
y_pred = automl.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
# RMSE 계산
rmse = mean_squared_error(y_test, y_pred, squared=False)
print("RMSE:", rmse)
```

```python
from sklearn.metrics import r2_score

# 실제 값과 예측 값 비교하여 R-squared 계산
r2 = r2_score(y_test, y_pred)
print("R-squared:", r2)
```

```python
import numpy as np

# 예측값과 실제값의 차이 계산
errors = np.abs(y_pred - y_test)

# 오차 범위 통계량 계산
mean_error = np.mean(errors)
max_error = np.max(errors)
min_error = np.min(errors)
median_error = np.median(errors)
std_error = np.std(errors)

print("Mean Error:", mean_error)
print("Max Error:", max_error)
print("Min Error:", min_error)
print("Median Error:", median_error)
print("Standard Deviation of Error:", std_error)
```

# 그러면.. 그.. 주성분분석을 하고 해보자

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# PCA 모델 생성 및 주성분 개수 설정
pca = PCA(n_components=2)

# 주성분 분석 수행
pca.fit(X)

# 주성분 변환
X_pca = pca.transform(X)

# 주성분의 설명된 분산 비율 확인
explained_variance_ratio = pca.explained_variance_ratio_
print("주성분의 설명된 분산 비율:", explained_variance_ratio)

plt.figure(figsize=(2, 1))
plt.scatter(X_pca[:,1], y)
plt.xlabel("주성분 1")
plt.ylabel("주성분 2")
plt.title("PCA result")
plt.show()
```

```python
from sklearn.decomposition import PCA
### PCA
# 주성분분석 함수화
def pcatotal(data, data2): # x는 독립변수 데이터
  pca = PCA(n_components=2)
  pca.fit(data)
  X_pca = pca.transform(data)
  print(pca.explained_variance_ratio_)
  print()

  df_pca = pd.DataFrame(X_pca, columns=['PCA1','PCA2'])
  df_pca['target']= phh['착과량(int)']

  plt.scatter(df_pca['PCA1'], df_pca['PCA2'], c=df_pca['target'])
  plt.xlabel('PCA1')
  plt.ylabel('PCA2')
  plt.colorbar(label='target')
  plt.show()

  # 주성분 1, 주성분 2, 연속형 종속변수 값 가져오기
  pc1 = df_pca['PCA1']
  pc2 = df_pca['PCA2']
  z = df_pca['target']

  # 3D 산점도 그래프 생성
  fig = plt.figure()
  ax = fig.add_subplot(111, projection='3d')

  # 산점도 그리기
  ax.scatter(pc1, pc2, z, c=z, cmap='rainbow')

  # 축 레이블 설정
  ax.set_xlabel('PCA1')
  ax.set_ylabel('PCA2')
  ax.set_zlabel('target')

  # 그래프 표시
  plt.show()
```
