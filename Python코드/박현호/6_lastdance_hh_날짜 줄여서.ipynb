```python
path = '/content/drive/MyDrive/ P - value/proj_3/데이터 파일/전처리데이터 (전후)/X_scaled.csv'

phh_X_scaled = pd.read_csv(path)
phh_X_scaled
```

```python
phh = phh_y_scaled.loc[:, '착과량(int)'].copy()
```

```python
phh = pd.concat([phh, phh_X_scaled_date.iloc[:, 0:14]], axis=1) # 새순 14일치
phh = pd.concat([phh, phh_X_scaled_date.iloc[:, 89:103]], axis=1) # 엽록소 14일치
```

```python
phh
```

```python
# X, y 나눠주기
X = phh.drop('착과량(int)', axis=1)  # 변수 열들을 X로 설정
y = phh['착과량(int)']  # 타겟 열을 y로 설정
```

```python
!pip install mljar-supervised
```

```python
from supervised.automl import AutoML
from sklearn.model_selection import train_test_split

# Train-Test 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)

# AutoML 모델 생성 및 학습
automl = AutoML(mode="Compete", eval_metric='rmse', algorithms=['Xgboost', 'Extra Trees', 'LightGBM'], stack_models=True)
automl.fit(X_train, y_train)

# 테스트 데이터에 대한 예측
y_pred = automl.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
# RMSE 계산
rmse = mean_squared_error(y_test, y_pred, squared=False)
print("RMSE:", rmse)
```

```python
from sklearn.metrics import r2_score

# 실제 값과 예측 값 비교하여 R-squared 계산
r2 = r2_score(y_test, y_pred)
print("R-squared:", r2)
```
